### P6. Write a Python code to output the following graph. (or a similar one). 
```

g1 = [80, 60, 80, 67, 34]

g3 = [30, 90, 50, 20, 90]
```

### P7. Write a Python code to output the following graph. (or a similar one)
```
Draw 25 dots. Generate random x and y coordinates (integers from 1 to 30), random colors and random sizes (integers from 20 to 500) for the dots.
```

```
Նկարեք 25 կետեր. Կետերի համար գեներացրեք պատահական x և y կոորդինատներ (1-ից 30 միջակայքում int տիպի թվեր), պատահական գույներ ու պատահական չափսեր (20-ից 500 միջակայքում int տիպի թվեր)։
```

````python
import numpy as np
import matplotlib.pyplot as plt
x1 = np.random.randint(30, size=(25))
y1 = np.random.randint(30, size=(25))
for x, y in zip(x1, y1):
    rgb = np.random.rand(3,)
    plt.scatter(x, y, color = [rgb], marker = "o", s = np.random.randint((20, 500)))
plt.xlim((0,30))
plt.ylim((0,30))
plt.title('Colorful dots')
plt.show()
````

```
Plot sepal_length and sepal_width features on the same histogram. 
```

````python
import numpy as np
import plotly.express as px
df = px.data.iris()
sepal_length = df['sepal_length']
sepal_width = df['sepal_width']


index = np.arange(len(sepal_length))
bar_width = 0.35

plt.bar(index - bar_width, sepal_length, bar_width, color='c', label='sepal_length')

plt.bar(index, sepal_width, bar_width, color='r', label='sepal_width')

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Sepal_length and sepal_width features')
plt.legend()


plt.show()
````

```
Create a heatmap showing the correlation of the columns sepal_width and petal_width.
```

````python
import plotly.figure_factory as ff
import numpy as np
import plotly.express as px
df = px.data.iris()
sepal_width = df['sepal_width']
petal_width = df['petal_width']

z = [sepal_width, petal_width]

fig = ff.create_annotated_heatmap(z)
fig.show()
````
